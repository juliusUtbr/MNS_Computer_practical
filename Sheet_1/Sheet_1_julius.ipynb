```python
import numpy as np

def McCullochPittsNeuron():
    """
    1. Simple McCulloch-Pitts Neuron

    -If you want more values for w and x you can add them to the arrays
    -Change values a and b to get all possible outputs for b)
    """

    a = 1
    b = 1

    # Some error handling
    if a != -1 and a != 1:
        print("Error: The value of a should be 0 or 1")
        exit()
    if b != -1 and b != 1:
        print("Error: The value of b should be 0 or 1")
        exit()

    # Initialize the weight vector and the input vector
    w = np.array([3, 2, 2])
    x = np.array([-1, a, b])

    # Some error handling
    if len(w) != len(x):
        print("Error: The length of the weight vector and the input vector should be the same")
        exit()
    if x[0] != -1:
        print("Error: The first element of the input vector should be -1")
        exit()

    # Calculate the output of the McCulloch-Pitts Neuron
    z = np.sign(np.dot(w, x))
    z = np.sign(z)

    return z

z = McCullochPittsNeuron()
print(z)



```

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoidFunction():

    a = [0.1,1,5]
    x = np.arange(-100, 100, 0.1)

    y = []

    for i in a:
        for j in x:
            y.append(2 / (1 + np.exp(i* j)) - 1)

        plt.plot(x, y, label="Sigmoid function: a = " + str(i))
        plt.xlabel("x")
        plt.ylabel("y", rotation=0)
        plt.legend()
        plt.show()

        y = []

def hyperbolicTangetFunction():

    a = [0.1,1,5]
    x = np.arange(-100, 100, 0.1)

    y = []

    for i in a:
        for j in x:
            y.append(np.tanh(i*j))
        
        plt.plot(x, y, label="Hyperbolic tangent function: a = " + str(i))
        plt.xlabel("x")
        plt.ylabel("y", rotation=0)
        plt.legend()
        plt.show()

        y = []

def piecewiseLinearFunction():

    a = [0.1,1,5]
    x = np.arange(-100, 10, 0.1)

    y = []

    for i in a:
        for j in x:

            if j >= 1/i:
                y.append(1)
            if -1/i < j < 1/i:
                y.append(i*j)
            if j <= -1/i:
                y.append(-1)

        plt.plot(x, y, label="Piecewise linear function: a = " + str(i))
        plt.xlabel("x")
        plt.ylabel("y", rotation=0)
        plt.legend()
        plt.show()

        y = []

def stepFunction():

    x = np.arange(-100, 100, 0.1)
    y = []

    for i in x:

        if i >= 0:
            y.append(1)
        else:
            y.append(0)
    
    plt.plot(x, y, label="Step function")
    plt.xlabel("x")
    plt.ylabel("y", rotation=0)
    plt.legend()
    plt.show()


print("Sigmoid function for three values of a")
sigmoidFunction()
print("\n")
print("Hyperbolic tangent function for three values of a")
hyperbolicTangetFunction()
print("\n")
print("Piecewise linear function for three values of a")
piecewiseLinearFunction()
print("\n")
print("Step function")
stepFunction()
```
